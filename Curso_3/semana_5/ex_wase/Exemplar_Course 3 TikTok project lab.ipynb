### **Task 4. Determine outliers**

Cuando se construyen modelos predictivos, la presencia de valores atípicos puede ser problemática. Por ejemplo, si estuvieras intentando predecir el número de vistas de un video en particular, los videos con un número extremadamente alto de vistas podrían introducir sesgos en el modelo. Además, algunos valores atípicos podrían indicar problemas con la forma en que se capturaron o registraron los datos.

El objetivo final del proyecto TikTok es construir un modelo que prediga si un video es una reclamación o una opinión. El análisis que has realizado indica que el nivel de participación de un video está fuertemente correlacionado con su estado de reclamación. No hay razón para creer que ninguno de los valores en los datos de TikTok esté capturado de manera errónea, y se ajustan a las expectativas de cómo funciona las redes sociales: una proporción muy pequeña de videos obtiene niveles de participación muy altos. Esa es la naturaleza del contenido viral.

Sin embargo, es una buena práctica tener una idea de cuántos de tus puntos de datos podrían considerarse valores atípicos. La definición de un valor atípico puede cambiar según los detalles de tu proyecto, y es útil tener experiencia en el dominio para decidir un umbral. Has aprendido que una forma común de determinar valores atípicos en una distribución normal es calcular el rango intercuartílico (IQR) y establecer un umbral que sea 1.5 * IQR por encima del tercer cuartil.

En este conjunto de datos de TikTok, los valores de las variables de conteo no siguen una distribución normal. Están sesgados hacia la derecha. Una forma de modificar el umbral de valores atípicos es calcular el valor **mediano** para cada variable y luego agregar 1.5 * IQR. Esto da como resultado un umbral que, en este caso, es mucho más bajo de lo que sería si se usara el tercer cuartil.

Escribe un bucle for que recorra los nombres de columna de cada variable de conteo. Para cada iteración:
1. Calcula el IQR de la columna.
2. Calcula la mediana de la columna.
3. Calcula el umbral de valores atípicos (mediana + 1.5 * IQR).
4. Calcula el número de videos con un conteo en esa columna que excede el umbral de valores atípicos.
5. Imprime "Número de valores atípicos, {nombre de la columna}: {conteo de valores atípicos}".
```
Example:
Number of outliers, video_view_count: ___
Number of outliers, video_like_count: ___
Number of outliers, video_share_count: ___
Number of outliers, video_download_count: ___
Number of outliers, video_comment_count: ___
```

````python
# Calcula el IQR de la columna.
iqr = data['video_view_count'].quantile(0.75) - data['video_view_count'].quantile(0.25)
# Calcula la mediana de la columna.
median = data['video_view_count'].median()
# Calcula el umbral de valores atípicos (mediana + 1.5 * IQR).
outlier_threshold = median + 1.5 * iqr
#Calcula el número de videos con un conteo en esa columna que excede el umbral de valores atípicos.
num_outliers = data[data['video_view_count'] > outlier_threshold].shape[0]
# Calcula el porcentaje de videos con un conteo en esa columna que excede el umbral de valores atípicos.
percent_outliers = num_outliers / data.shape[0]
# Imprime los resultados.
print('IQR: ', iqr)
print('Mediana: ', median)
print('Umbral de valores atípicos: ', outlier_threshold)
print('Número de valores atípicos: ', num_outliers)
print('Porcentaje de valores atípicos: ', percent_outliers)
````

````python
count_cols = ['video_view_count',
              'video_like_count',
              'video_share_count',
              'video_download_count',
              'video_comment_count',
              ]

for column in count_cols:
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    median = data[column].median()
    outlier_threshold = median + 1.5*iqr

    # Count the number of values that exceed the outlier threshold
    outlier_count = (data[column] > outlier_threshold).sum()
    print(f'Número de valores atípicos en  {column:<21}:', outlier_count)
````
